# Ex1: Régression linéaire et intervalle de confiance

$$\mathop J(\Theta)=(X^\top\Theta-y)\top . (X^\top\Theta -y)$$

La forme de la fonction est: $$\mathop f*g$$  
La dérivée de $$\mathop (f*g)' = f'g +fg' $$ 
On a donc
$$\mathop f' = [(X^\top\Theta-y)\top]' = X $$
$$\mathop g' = [(X^\top\Theta -y)]' = X\top $$ 
Du coup on a
$$\mathop : \frac{\partial J}{\partial \Theta} (\Theta) = X(X^\top\Theta -y) + X\top(X^\top\Theta-y)\top = 2X(X^\top \Theta - y)$$ 


\begin{align*}
\frac{\partial J}{\partial \Theta} (\Theta) = 2X(X^\top \Theta - y)
\end{align*}
On cherche ou ça s'annule
$$\mathop 2X(X^\top\Theta - y)=0$$
$$\mathop X(X^\top\Theta - y)=0$$
$$\mathop XX^\top\Theta - Xy=0$$
$$\mathop XX^\top\Theta=Xy$$
$$\mathop \Theta=Xy / XX^\top$$

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
X = np.linspace(0,5,num=90)
print(X.shape)
print(X)
```

```python
mu = 2*X + 1
sigma = 0.2
y = np.random.normal(mu, sigma, 90)
print(y.shape)
print(y)
```

```python
plt.scatter(X, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
print(X_train.shape)
print(X_train)
```

```python
X_train = np.vstack((X_train, np.ones(len(X_train))))
print(X_train.shape)
print(X_train)
```

```python
theta = np.linalg.inv(X_train@X_train.T)@X_train@y_train
print(theta)
```

```python
plt.scatter(X_train[:-1], y_train, c="b")
plt.scatter(X_test, y_test, c="r")
plt.plot(X, theta[0]*X + theta[1], c="y")
```

```python
sig = np.sqrt(1/90 * np.sum( (y - theta[0]*X - theta[1])**2 ))
print(sig)
```

```python
plt.scatter(X_train[:-1], y_train, c="b")
plt.scatter(X_test, y_test, c="r")
plt.plot(X, theta[0]*X + theta[1], c="g")
plt.plot(X, theta[0]*X + theta[1] - 3*sig, c="y")
plt.plot(X, theta[0]*X + theta[1] + 3*sig, c="y")
```

# Ex2 : Régression polynomiale et régularisation Ridge

```python
X2 = np.linspace(-10,10,num=90)
print(X2.shape)
print(X2)

mu = np.sin(X2)/X2
sigma = 0.05
y2 = np.random.normal(mu, sigma, 90)
print(y2.shape)
print(y2)

plt.scatter(X2,y2)
```

```python
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.33, random_state=42)
#X2_train = np.vstack((X2_train, np.ones(len(X2_train))))
print(X2_train.shape)
print(X2_train)
```

```python
def polyreg(X_train, y_train, d):
    z = np.ones(len(X_train))
    for i in range(1, d+1):
        z = np.vstack((z, np.power(X_train, i)))
    theta = np.linalg.inv(z@z.T)@z@y_train
    return theta, z
```

```python
theta, z = polyreg(X2_train, y2_train, 3)
c = theta.reshape(-1,1) * z
c = c.sum(axis=0)

plt.scatter(X2,y2)
plt.scatter(X2_train, c)
```

```python
theta, z = polyreg(X2_train, y2_train, 8)
c = np.sum(theta.reshape(-1,1) * z, axis=0)

plt.scatter(X2,y2)
plt.scatter(X2_train, c)
```

```python
for i in range(1,28):
    theta_train, z_train = polyreg(X2_train, y2_train, i)
    y_pred_train = np.sum(theta_train.reshape(-1,1) * z_train, axis=0)
    err_train = np.sum((y_pred_train - y2_train)**2)
    
    theta_test, z_test = polyreg(X2_test, y2_test, i)
    y_pred_test = np.sum(theta_test.reshape(-1,1) * z_test, axis=0)
    err_test = np.sum((y_pred_test - y2_test)**2)
    
    plt.scatter(i, err_train, c='r')
    plt.scatter(i, err_test, c='b')
```

```python
theta, z = polyreg(X2_train, y2_train, 19)
c = np.sum(theta.reshape(-1,1) * z, axis=0)

plt.scatter(X2,y2)
plt.scatter(X2_train, c)
```

On détecte l'overfitting car les pred collent beaucoup trop à notre donnée.

```python
def polyreg_ridge(X_train, y_train, d, sig):
    z = np.ones(len(X_train))
    for i in range(1, d+1):
        z = np.vstack((z, np.power(X_train, i)))
    theta = np.linalg.inv(z@z.T + np.identity(len(z)) * sig)@z@y_train
    return theta, z
```

```python
theta, z = polyreg_ridge(X2_train, y2_train, 19, 1)
c = np.sum(theta.reshape(-1,1) * z, axis=0)

plt.scatter(X2,y2)
plt.scatter(X2_train, c)
```
