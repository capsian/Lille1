```python
import torch
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
X = iris['data']
y = iris['target']
names = iris['target_names']
feature_names = iris['feature_names']

print(names)
print(feature_names)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=76)
```

```python
X_train = torch.tensor(X_train.astype(np.float32))
X_test = torch.tensor(X_test.astype(np.float32))
y_train = torch.tensor(y_train.astype(np.int))
y_test = torch.tensor(y_test.astype(np.int))
```

```python
plt.figure(figsize=(10,10))

loss_fn = torch.nn.NLLLoss()

learning_rates = [1e-4, 1e-2, 5e-1, 1, 2]
for lr in learning_rates:
    loss_descent = []
    model = torch.nn.Sequential(
        torch.nn.Linear(4, 3),
        torch.nn.ReLU(),
        torch.nn.Linear(3, 3),
        torch.nn.Softmax(dim=1)
    )

    for t in range(10000):
        y_pred = model(X_train) # forward pass
        loss = loss_fn(y_pred, y_train)

        if t % 100 == 99:
            #print(t, loss.item())
            loss_descent.append(loss.item())

        model.zero_grad()
        loss.backward()
        with torch.no_grad():
            for param in model.parameters():
                param -= lr * param.grad

    plt.plot(loss_descent, label=str(lr))
    
plt.legend()
plt.show()
```
