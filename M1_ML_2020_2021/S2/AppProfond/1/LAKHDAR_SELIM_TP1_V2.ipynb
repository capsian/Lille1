```python
import numpy as np
import math
import matplotlib.pyplot as plt

x = np.linspace(-math.pi, math.pi, 2000)
y = np.sin(x)
plt.scatter(x,y)
```

```python
a = np.random.rand()
b = np.random.rand()
c = np.random.rand()
d = np.random.rand()

print("initial a: " + str(a))
print("initial b: " + str(b))
print("initial c: " + str(c))
print("initial d: " + str(d))

y_pred = a + b*x + c*x*x + d*x*x*x

plt.scatter(x,y)
plt.plot(x,y_pred, "r")

nb_iter = 2000
learning_rate = 0.001
for i in range(nb_iter):
    y_pred = a + b*x + c*x*x + d*x*x*x

    d_a = np.sum((y_pred - y)*2*1) * 1/len(x)
    d_b = np.sum((y_pred - y)*2*x) * 1/len(x)
    d_c = np.sum((y_pred - y)*2*x*x) * 1/len(x)
    d_d = np.sum((y_pred - y)*2*x*x*x) * 1/len(x)
    
    a = a - (learning_rate * d_a)
    b = b - (learning_rate * d_b)
    c = c - (learning_rate * d_c)
    d = d - (learning_rate * d_d)
```

```python
y_pred_new = a + b*x + c*x*x + d*x*x*x

plt.plot(x,y)
plt.plot(x,y_pred_new, "r")

print("final a: " + str(a))
print("final b: " + str(b))
print("final c: " + str(c))
print("final d: " + str(d))
```
