**Question** Créer  4 classifieurs de classe suivante. On prendra 5 voisins pour le $k$-plus proches voisins, et une profondeur de 20 pour les arbres de décision. Les mémoriser dans une liste `methods` pour plus de facilité.  

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
svc = SVC()
sgd = SGDClassifier()
lr = LogisticRegression()
dt = DecisionTreeClassifier(max_depth=20)
knn = KNeighborsClassifier(5)
methods = [gnb, dt, sgd, knn, lr, svc]
```

On considère les tailles de 100 à 4100 par pas de 1000. Par la suite on prendra des sous-échantillon de `X_train` tiré au hasard des tailles considérées (vous jouerez sur les indices plutôt que sur les données!!)

    sizes = list(range(100, 5000, 1000)) 

```python
sizes = list(range(100, 5000, 1000)) 
```

# Mesurer les temps d'exécution

**Question** Regardez la documentation de %timeit

```python
%timeit?
```

**Question** Que fait l'instruction ? 

    r = %timeit -o -q -r1 -n1 mon_appel_de_fonction

**Question** Remarquez ce que donne 
  
    <instance_de_methode_de_classif>.__class__.__name__

**Question** En utilisant cela, faites le diagramme suivant :

![temps d'exécution](./temps.png)

Dans chaque courbe, les modèles sont entraînés sur les mêmes données, qui sont des sous échantillons de `X_train` tirés au hasard de chaque taille de la liste `sizes`. 

```python
tps_fit = {}
tps_pred = {}
for m in methods:
    tps_fit[m] = []
    tps_pred[m] = []
for size in sizes:
    indices = np.random.choice(list(range(len(X_train))), size)
    for m in methods:
        tf = %timeit -o -q -r1 -n1 m.fit(X_train[indices], y_train[indices])
        tp = %timeit -o -q -r1 -n1 m.predict(X_train[indices])
        tps_fit[m].append(tf.timings[0])
        tps_pred[m].append(tp.timings[0])
        print("{:>25} ({}) fit : {:1.3f} ; Pred {:1.3f}".format(m.__class__.__name__, 
                                                                size, 
                                                                tf.timings[0],
                                                                tp.timings[0]))
    print("-"*20)                 
```

```python
fig, ax = plt.subplots(2, figsize=(5, 10), sharex = True)
for m in methods:
    ax[0].plot(sizes, tps_fit[m], label=m.__class__.__name__)
    ax[1].plot(sizes, tps_pred[m], label=m.__class__.__name__)
ax[0].legend()
ax[1].legend()
ax[0].set_xlabel("taille")
ax[1].set_ylabel("temps (s)")
ax[0].set_ylabel("temps (s)")
None
```

# Mesurer les scores

**Question** Importer cross_val_score et lire la documentation

```python
from sklearn.model_selection import cross_val_score
```

**Question** Tracer les courbes suivantes obtenues avec cross_val_score sur les différentes tailles. 

![precision](./acc.png)

(et vous voulez, car c'est plus joli... :-) )

![precision enveloppe](./acc_b.png)

```python
acc = {}
std = {}
for m in methods:
    acc[m] = []
    std[m] = []
for size in sizes:
    indices = np.random.choice(list(range(len(X_train))), size)
    for m in methods:
        r = cross_val_score(m, X_train[indices], y_train[indices])
        acc[m].append(np.mean(r))
        std[m].append(np.std(r))
        print("{:>25} ({}): {:1.3f} ({:1.3f})".format(m.__class__.__name__, size, np.mean(r), np.std(r)))
    print('-'*20)
```

```python
fig, ax = plt.subplots()
for m in methods:
    ax.errorbar(sizes, acc[m], yerr=std[m], label=m.__class__.__name__)
ax.legend()
ax.set_xlabel("taille")
ax.set_ylabel("Précision")
None
```

```python
fig, ax = plt.subplots()
for m in methods:
    ax.plot(sizes, acc[m], label=m.__class__.__name__)
    ax.fill_between(sizes, np.array(acc[m])-np.array(std[m]), np.array(acc[m])+np.array(std[m]), alpha=.2)
ax.legend()
ax.set_xlabel("taille")
ax.set_ylabel("Précision")
None
```

# Matrice de confusion

On change de problème de classification en un problème binaire de détermination si un digit est le '3'.

```python
y_train3 = (y_train == '3')
y_test3 = (y_test == '3')
```

**Question** Sélectionnez 5000 indices au hasard de X_train pour travailler sur un sous échantillon (nommé `Xs`, `ys`). Découpez ce sous-échantillon pour avoir un échantillon d'apprentissage et de validation. 

```python
from sklearn.model_selection import train_test_split
indices = np.random.choice(range(len(y_train3)), 5000)
```

```python
Xs, ys = X_train[indices], y_train3[indices]
Xs_train, Xs_valid, ys_train, ys_valid = train_test_split(Xs, ys)
```

**Question** Entraînez tous les classifieurs sur ce sous-échantillon et calculer dans un dictionnaire `pred` toutes les prédictions de chaque méthode sur l'échantillon de validation. 

```python
pred = {}
for m in methods:
    m.fit(Xs_train, ys_train)
    pred[m] = m.predict(Xs_valid)
```

**Question** Calculer les matrices de confusion pour `lr` et `sgd`. Affichez aussi les matrices de confusion de façon graphique (`plot_confusion_matrix`)


```python
from sklearn import metrics 
```

```python
print(metrics.confusion_matrix(pred[lr], ys_valid))
```

```python
print(metrics.confusion_matrix(pred[sgd], ys_valid))
```

```python
fig, ax = plt.subplots(1,2, gridspec_kw={'hspace':1, 'wspace':1})
metrics.plot_confusion_matrix(lr, Xs_valid, ys_valid, ax=ax[0])
metrics.plot_confusion_matrix(sgd, Xs_valid, ys_valid, ax=ax[1])

ax[0].set_title("Logistic Reg")
ax[1].set_title("SGD")
None
```
